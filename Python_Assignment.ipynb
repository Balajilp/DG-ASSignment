Define a class Person and its two child classes: Male and Female. All classes have a
method "get_gender" which can print "Male" for Male class and "Female" for Female
Class.

Bonus: Make class Person an abstract class and make get_gender an abstract method in the same class. The two child classes must inherit and implement get_gender. i.,e, When trying to initialize an object of class Person, the program must throw an error.
Hint: 
Use abc library (comes natively with Python3) https://www.python-course.eu/python3_abstract_classes.php

```python
from abc import ABC, abstractmethod

class person(ABC):

    @abstractmethod
    def get_gender(self):
        pass


class Male(person):
    def get_gender(self):
        print('Male')


class Female(person):
    def get_gender(self):
        print('Female')

object = Male()
object.get_gender()
```

# Quesion 3
With a given list [12,24,35,24,88,120,155,88,120,155], write a program to print this list
after removing all duplicate values with original order reserved.
Hint: Use set() to store a number of values without duplicates.


```python
# python function to remove duplicates from the give list
# we can use set method also but the problem is set is unordered.  we can't able to preserve the original order
def remove_duplicates(lst):
    result = []
    for value in lst:
        if value not in result:
            result.append(value)

    print(result)

lst = list(map(int, input().split())) # input : 1 2 3 4 5
remove_duplicates(lst)
```

# Question 5
Write a binary search function which searches an item in a sorted list. The function
should return the index of the element to be searched in the list


```python
def BinarySearch(list1, key):
    low = 0
    high = len(list1)-1
    Found = 0
    while low <= high and not Found:
        mid = (low+high)//2
        if key == list1[mid]:
            Found = mid
        elif key > list1[mid]:
            low = mid+1
        else:
            high = mid-1
    if Found:
        print(Found)
    else:
        print('Key is not found')

list1 = list(map(int, input().split()))
list1.sort()
key = int(input())
BinarySearch(list1, key)
```

# Question 6
Write a program using generator to print the numbers which can be divisible by 5
and 7 between 0 and n in comma separated form while n is input by console.


```python
def generator(n):
  for x in range(n):
    if x%5==0 and x%7==0:
      yield x 

n = int(input())
value = generator(n)

for x in value:
  print(x)
```

# Question 7
Assuming that we have some email addresses in the "username@companyname.com"
format, write a program to print the company name of a given email address. Both user
names and company names are composed of letters only.


```python
def check_email(email):
    res = ''
    for x in range(len(email)):
        if email[x] == '@':
            for y in range(x+1, len(email)):
                if email[y] == '.':
                    break
                else:
                    res += email[y]
    print(res)

email = input() # input : username@company.com
check_email(email)
```

in that directory recursively as a tree. 
Hint: Use os.listdir and os.path.isdir functions

```python
import os
def get_files(path):
    
    try:
        files = os.listdir(path)

        for f in files:
            print(f)
    except:
        print('Enter the valid path')
        get_files(path)
path = input()
get_files(path)
```

# Question 11
Write a program anti_html.py that takes a URL as argument, downloads the html from web
0and print it after stripping html tags.


```python
import urllib.request, urllib.error, urllib.parse
import re

def get_url(url):
    response = urllib.request.urlopen(url)
    webContent = response.read().decode('UTF-8')

    TAG_RE = re.compile(r'<[^>]+>')
    def remove_tags(text):
        return TAG_RE.sub('', text)


    ob = remove_tags(webContent)
    print(ob)

url = input()

get_url(url)
```
